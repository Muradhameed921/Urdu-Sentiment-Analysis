```python
#Reading the file we loaded before and spliting by new lines to create the list of stopwords.
#These stopwords will not contain any meaning and hence we have to remove them
file_name = 'filtered_stopwords.txt'
with open(file_name, 'r', encoding='utf-8') as file:
    stopwords = [line.strip() for line in file.readlines()]
print(stopwords)
```

```python
#Defining a custom function that will remove the stopwords
def remove_stopwords(text):
  if isinstance(text, str):
    new_text = []
    for word in text.split():
      if word in stopwords:
        new_text.append('')
      else:
        new_text.append(word)
    return " ".join(new_text)
  else:
    return text
```

```python
#Applying to the dataset
print(df['urdu_text'][656])
print(remove_stopwords(df['urdu_text'][656]))
df['urdu_text']=df['urdu_text'].apply(remove_stopwords)
df.head(10)
```

# **Punctuation, Emojis, and Hashtags**

```python
#Removing urls by creating custom function and using REGEX from the dataset as they do not add to the semantic meaning

import re

def remove_url(text):
    if isinstance(text, str):
        pattern = re.compile(r'https?://\S+|www\.\S+')
        return pattern.sub(r'', text)
    else:
        return text
print(df['urdu_text'][5619])
print(remove_url(df['urdu_text'][5619]))
df['urdu_text']=df['urdu_text'].apply(remove_url)
df.head(10)
```

```python
#Removing hashtags by creating custom function and using REGEX from the dataset as they do not add to the semantic meaning
def remove_hashtags(text):
    if isinstance(text, str):
      return re.sub(r'#\S+', '', text)
    else:
      return text
print("میں آج بہت خوش ہوں #خوش")
print(remove_hashtags("میں آج بہت خوش ہوں #خوش"))
df['urdu_text']=df['urdu_text'].apply(remove_hashtags)
df.head(10)
```

### **TF-IDF Vectorization**

```python
#Perfroming TFIDF vectorization and fetching the 10 most rare words in the whole corpus

import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from collections import Counter


vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df['urdu_text'])
feature_names = vectorizer.get_feature_names_out()
tfidf_scores = X.toarray().sum(axis=0)
corpus_tokens = [word for doc in df['urdu_text'] for word in doc.split()]
word_counts = Counter(corpus_tokens)
word_frequencies = [word_counts[word] if word_counts[word] > 0 else 1 for word in feature_names]
adjusted_tfidf_scores = tfidf_scores / word_frequencies
results = pd.DataFrame({'word': feature_names, 'tfidf': adjusted_tfidf_scores})
results = results.sort_values(by='tfidf', ascending=False)
print(results.head(10))
```

# **N-Grams**

```python
#Making tokenns of the dataset i.e 1 token at a time, 2 tokens at a time, 3 tokens at a time

import pandas as pd
from nltk import bigrams, trigrams
from collections import Counter

def urdu_tokenize(sentence):
    tokens = urdu_text_processing.urdu_tokenize(sentence)
    return tokens

unigram_counter = Counter()
bigram_counter = Counter()
trigram_counter = Counter()

for sentence in df['urdu_text']:
    tokens = urdu_tokenize(sentence)
    unigram_counter.update(tokens)
    bigram_counter.update(bigrams(tokens))
    trigram_counter.update(trigrams(tokens))

top_bigrams = bigram_counter.most_common(10)
top_trigrams = trigram_counter.most_common(10)

print("Top 10 Bigrams:")
for bigram, freq in top_bigrams:
    print(f"{bigram}: {freq}")

print("\nTop 10 Trigrams:")
for trigram, freq in top_trigrams:
    print(f"{trigram}: {freq}")
```

### **Logistic Regression Using Word2Vec**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from gensim.models import Word2Vec
import re


def clean_and_tokenize(text):
    text = re.sub(r'[^\w\s]', '', text)
    return text.split()
tokenized_texts = df['urdu_text'].apply(clean_and_tokenize)
word2vec_model = Word2Vec(sentences=tokenized_texts.tolist(), vector_size=100, window=5, min_count=1, workers=4)
def get_average_vector(tokens):
    valid_tokens = [token for token in tokens if token in word2vec_model.wv.key_to_index]
    if valid_tokens:
        return np.mean(word2vec_model.wv[valid_tokens], axis=0)
    else:
        return np.zeros(word2vec_model.vector_size)
X_word2vec = np.array([get_average_vector(tokens) for tokens in tokenized_texts])
y = df['is_sarcastic'].values
X_train, X_test, y_train, y_test = train_test_split(X_word2vec, y, test_size=0.2, random_state=42)
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred, target_names=['Not Sarcastic', 'Sarcastic']))
```

## **Naive Bayes with word2vec**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report
from gensim.models import Word2Vec
import numpy as np

sentences = [text.split() for text in df['urdu_text']]
word2vec_model = Word2Vec(sentences, vector_size=100, window=5, min_count=1, workers=4)
def get_avg_word2vec(sentence, model):
    vectors = [model.wv[word] for word in sentence if word in model.wv]
    if len(vectors) > 0:
        return np.mean(vectors, axis=0)
    else:
        return np.zeros(model.vector_size)

X = np.array([get_avg_word2vec(sentence, word2vec_model) for sentence in sentences])
y = df['is_sarcastic']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
model = GaussianNB()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred, target_names=['Not Sarcastic', 'Sarcastic']))
```
